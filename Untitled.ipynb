# Used card price prediction

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# pd.set_option("display.max_rows",None)
```

```python
# reading the csv with pandas
vehicles = pd.read_csv("datasets/vehicles.csv")
```

```python
# csv info
vehicles.info()
```

```python
# verify missing values 
vehicles.isnull().sum()
```

```python
# after reviewing the data we will delete the following columns:
# url, region, region_url, VIN, image_url, description, county, lat, long, posting_date
# this columns some of them contain data that can't be use and don't have an impact in our model in anyways
vehicles.drop(['url','region','region_url','VIN','image_url','description','county','lat','long', 'posting_date'],axis=1,inplace=True)
```

```python
vehicles.info()
```

```python
vehicles.isnull().sum()
```

```python
vehicles['price'].describe()
vehicles_originala = vehicles.copy()
```

```python
vehicles = vehicles[(vehicles['price']>1000)&(vehicles['price']<150000)]

plt.hist(vehicles['price'], bins=50, color='blue',edgecolor='black')
plt.xlabel("pret")
plt.ylabel("frecventa pretului")
plt.show()
```

```python
vehicles['price'].describe()
```

```python
vehicles.dropna(subset=['year','model','fuel','odometer','title_status','transmission'],inplace=True)
```

```python
vehicles.isnull().sum()
```

```python
vehicles.info()
```

```python
plt.hist(vehicles['price'],bins=150,color='lightblue',edgecolor='black')
plt.xlabel('Preț')
plt.ylabel('Număr vehicule')
plt.title('Distribuția prețurilor (histogramă)')
plt.grid(axis='y',linestyle='--', alpha=0.7)
plt.show()
```

```python
vehicles['year'].describe()
```

```python
vehicles['year'].value_counts()
```

```python
vehicles_year = vehicles[(vehicles['year']>=1970)&(vehicles['year']<=2022)]
plt.hist(vehicles_year['year'],bins=52,color='green', edgecolor='black')
plt.xlabel('Ani')
plt.ylabel('Distributia masinilor pe anii de fabricatii')
plt.grid(axis='y',linestyle='--')
plt.show()
```

```python
vehicles['manufacturer'].value_counts().head(100).plot(kind='bar')
plt.xlabel('Brand')
plt.ylabel('Numar de vehicule')
plt.grid(axis='y', linestyle='--')
plt.show()
```
